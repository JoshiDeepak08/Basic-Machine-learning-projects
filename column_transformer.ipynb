```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data.csv")
data
```

```python
df = data.drop(columns= [ 'Difficulty Breathing','Fatigue','Cholesterol Level','Fever', 'Outcome Variable']).values
```

```python
df
```

```python
DF= pd.DataFrame(df, columns=[ 'Disease','Fever','Age','Gender', 'Blood pressure'])
```

```python
DF
```

```python
# disease--> one hot encoding
# Blood pressure--> ordinal incoding
# gender and fever--> one hot encoding
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(strategy='most_frequent'), ['Fever']),
    ('tnf2', OrdinalEncoder(), ['Blood pressure']),
    ('tnf3', OneHotEncoder(sparse=False, drop="first"), ['Disease', 'Gender'])
], remainder='passthrough')
```

```python
transformer.fit_transform(DF).shape
```
